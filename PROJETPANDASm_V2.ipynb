```python
import pandas as pd
import matplotlib.pyplot as plt

# sample data
raw_data = {'pays': ['Europe', 'Moyen-Orient', 'Amérique centrale et du Sud', 'UE27', 'Eurasie', 'Afrique', 'UE28', 'Asie Pacifique', 'Amérique du Nord'],
        'consommation_gm3_an': [ 7458.8, 6374.4, 2182.0, 396.6, 7365.9,  1659.5, 5700.5, 8962.9, 12190.4  ],
        'annee': [0,  2008,2010,2012,2014, 2016, 2018,2020,2021]}
df = pd.DataFrame(raw_data, columns = ['pays', 'consommation_gm3_an', 'annee'])

# calculate percentage of each country's consumption
df['pourcentage'] = df['consommation_gm3_an'] / df['consommation_gm3_an'].sum() * 100



# plot chart
plt.figure(figsize=(18,9))
plt.pie(df['pourcentage'], labels=df['pays'], autopct='%1.1f%%', startangle=90, textprops={'fontsize': 15})
plt.title("Répartition de la consommation de l'energie entre 2008 et 2021",fontsize=20)
plt.legend(fontsize=13,bbox_to_anchor=(1.10, 1))
plt.show()
```
